```python
# Introduction to "list" in python
list1=["apple","banana","orange"]
print(list1)
```

```python
# To print Alphanumeric list
list2 = [2,2.3,"pranav",True]
print(list2)
```

```python
# To implement Indexing of the list
list4=["apple","banana","orange","cherry","mango","kiwi"]
print(list4[2])
print(list4[-4])
print(list4[2:5])
print(list4[:4])
print(list4[2:])
```

```python
# To replace an element in list
list5=["apple","orange","mango"]
list5[2] = "cherry"
print(list5)
```

```python
# To add an element in the list
list6=["apple","orange","cherry"]
list6.append("mango")
print(list6)
```

```python
# To change the position of an element in list
list7=["apple","banana","orange",]
list7.insert(1,"mango")
print(list7)
```

```python
# To combine two list
list8=["apple","banana","orange"]
list9=["mango","kiwi","cherry"]
list8.extend(list9)
print(list8)
print(list9)
```

```python
# Practice Questions
marks = [90,82,66,76,33]
print("The Maximum Marks are : " , max(marks))
print("The Minimum Marks are : " , min(marks))
print("The Number of Students are : " , len(marks))
print("The Sum of marks of students are : " , sum(marks))
print("The Average marks of students are : " , sum(marks)/len(marks))
marks.sort()
print("Sorted array of marks are : " , (marks))
print("Sorted array of marks are : " , sorted(marks))
```
